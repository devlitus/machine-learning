## 1. Propuesta de Sistema de IA de Alto Riesgo

### Descripción del sistema

El sistema propuesto es un **asistente de triaje médico inteligente** para servicios de urgencias hospitalarias. Utiliza **Procesamiento del Lenguaje Natural (NLP)** para analizar la descripción verbal o escrita de los síntomas del paciente, combinada con datos de constantes vitales, y proporciona una **clasificación de urgencia** (p. ej. escala Manchester: rojo, naranja, amarillo, verde, azul).

**Objetivo:** Reducir los errores de priorización, disminuir tiempos de espera críticos y apoyar la decisión del personal sanitario en urgencias, especialmente en situaciones de alta demanda.

---

### Clasificación según el Reglamento de IA de la UE — Anexo III

El sistema se encuadra en el **punto 5.d del Anexo III**:

> *"Sistemas de IA destinados a ser utilizados para la evaluación y la clasificación de las llamadas de emergencia realizadas por personas físicas o para el envío o el establecimiento de prioridades en el envío de servicios de primera intervención en situaciones de emergencia [...] y en **sistemas de triaje de pacientes en el contexto de la asistencia sanitaria de urgencia**."*

Por tanto, se trata de un **sistema de IA de alto riesgo** según el artículo 6.2 del Reglamento.

## 2. Validación: No es Riesgo Inaceptable

### ¿Por qué es **Alto Riesgo** y no **Riesgo Inaceptable**?

Según el Artículo 5 del Reglamento de IA de la UE, las prácticas de **riesgo inaceptable** son aquellas que:
- Manipulan subconscientemente el comportamiento humano causando daño
- Explotan vulnerabilidades de grupos específicos (edad, discapacidad)
- Permiten la puntuación social por autoridades públicas
- Usan sistemas de identificación biométrica remota en tiempo real en espacios públicos con fines policiales (con excepciones muy limitadas)

Nuestro sistema de triaje **NO incurre en ninguna de estas prácticas**, ya que:

1. **No manipula** a los pacientes: es una herramienta de apoyo a la decisión clínica
2. **No discrimina** por características sensibles; clasifica por urgencia médica objetiva
3. **No elabora perfiles generales** de personas: actúa sobre el episodio médico concreto
4. **Siempre hay supervisión humana**: el enfermero/a toma la decisión final, la IA solo recomienda
5. **No opera en tiempo real en espacios públicos** con fines de vigilancia o control social

El sistema sí es **Alto Riesgo** (Art. 6.2 + Anexo III, punto 5.d) porque su salida puede influir sustancialmente en decisiones que afectan la salud y la vida de personas físicas.

## 3. CRISP-DM Fase 1 — Objetivos del Negocio

### 3.1 ¿Cuál es el problema que se desea resolver?

Los servicios de urgencias hospitalarias padecen una **sobrecarga crónica** que genera errores en la priorización de pacientes (triaje). El proceso manual depende de la experiencia del enfermero/a, está sujeto a fatiga, variabilidad humana y falta de tiempo. Esto puede derivar en:

- Retrasos en la atención a pacientes críticos
- Infrautilización de recursos ante casos menores con alta presión percibida
- Alta tasa de mortalidad evitable por errores de priorización

**El sistema de IA propone apoyar (no reemplazar) al profesional**, ofreciendo una clasificación basada en datos objetivos que reduzca la variabilidad y mejore la velocidad de respuesta.

---

### 3.2 ¿Por qué utilizar Inteligencia Artificial?

| Razón | Justificación |
|---|---|
| **Volumen de datos** | Miles de episodios diarios en hospitales grandes; imposible revisar manualmente todos |
| **Patrones complejos** | Los síntomas tienen combinaciones no lineales que los modelos ML capturan mejor que reglas fijas |
| **NLP para texto libre** | Los pacientes describen síntomas en lenguaje natural; NLP (BERT médico) permite extraer entidades clínicas automáticamente |
| **Tiempo real** | El modelo infiere en milisegundos, crítico en urgencias |
| **Aprendizaje continuo** | El modelo mejora con nuevos casos sin reescribir reglas |

**Tecnología específica:** Clasificación multiclase supervisada con fusión de modelos (NLP + datos tabulares de constantes vitales), usando arquitectura Transformer para el texto y XGBoost para los datos numéricos.

---

### 3.3 Criterios de Éxito

```python
import numpy as np

# Definición de criterios de éxito cualitativos y cuantitativos
criterios = {
    "Criterio": [
        "Precisión del modelo (accuracy)",
        "Reducción de errores de priorización",
        "Tiempo medio de triaje",
        "Aceptación del personal sanitario",
        "Cobertura de casos atendidos por el sistema",
        "Conformidad con la Ley de IA UE",
        "Explicabilidad de las decisiones"
    ],
    "Tipo": [
        "Cuantitativo", "Cuantitativo", "Cuantitativo",
        "Cualitativo", "Cuantitativo", "Cualitativo", "Cualitativo"
    ],
    "Meta / Indicador": [
        "≥ 90% en conjunto de prueba (F1-score macro)",
        "Reducción ≥ 30% vs. triaje manual en auditoría",
        "≤ 2 minutos por paciente (vs. 5-8 min actual)",
        "≥ 80% de enfermeros/as valoran el sistema como útil",
        "≥ 95% de episodios procesados sin error técnico",
        "Documentación, registro y supervisión humana activa",
        "Cada predicción incluye justificación en lenguaje natural"
    ]
}

df_criterios = pd.DataFrame(criterios)

# Visualización: radar/bar de métricas cuantitativas
metricas = ["Precisión\n(objetivo 90%)", "Reducción errores\n(objetivo 30%)",
            "Cobertura\n(objetivo 95%)", "Aceptación\npersonal (80%)"]
objetivos = [90, 30, 95, 80]
colores = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12']

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Gráfico de barras con objetivos
bars = axes[0].barh(metricas, objetivos, color=colores, edgecolor='white', height=0.5)
axes[0].set_xlim(0, 105)
axes[0].set_xlabel("Umbral mínimo de éxito (%)", fontsize=10)
axes[0].set_title("Criterios de Éxito Cuantitativos", fontsize=11, fontweight='bold')
for bar, val in zip(bars, objetivos):
    axes[0].text(val + 1, bar.get_y() + bar.get_height()/2, f'{val}%',
                 va='center', fontsize=10, fontweight='bold')
axes[0].axvline(x=80, color='gray', linestyle='--', alpha=0.5, label='Mínimo aceptable 80%')
axes[0].legend(fontsize=8)

# Tabla de criterios completos
axes[1].axis('off')
table2 = axes[1].table(
    cellText=df_criterios.values,
    colLabels=df_criterios.columns,
    cellLoc='left',
    loc='center',
    colWidths=[0.3, 0.15, 0.45]
)
table2.auto_set_font_size(False)
table2.set_fontsize(7.5)
table2.scale(1, 1.8)
for (row, col), cell in table2.get_celld().items():
    if row == 0:
        cell.set_facecolor('#2c3e50')
        cell.set_text_props(color='white', fontweight='bold')
    elif df_criterios["Tipo"].iloc[row-1] == "Cuantitativo" if row > 0 else False:
        cell.set_facecolor('#d6eaf8')
    else:
        cell.set_facecolor('#fdfefe')

plt.suptitle("CRISP-DM: Criterios de Éxito del Proyecto", fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('criterios_exito.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. CRISP-DM Fase 1 — Evaluación de la Situación

### 4.1 Conocimiento previo disponible sobre el problema

El triaje médico es un campo ampliamente estudiado. Existen sólidas bases de conocimiento:

- **Escala de Triaje de Manchester (MTS):** estándar europeo con 5 niveles de urgencia y más de 50 flujos clínicos documentados
- **Literatura científica:** más de 2.000 publicaciones en PubMed sobre sistemas de IA aplicados a triaje (2015-2024)
- **Modelos existentes:** proyectos como *DeepTriaje* (Hospital La Fe, Valencia) y trabajos del MIT sobre NLP en urgencias
- **Historial clínico electrónico (HCE):** los hospitales modernos disponen de registros digitales que pueden usarse para entrenamiento

> El dominio está bien definido y existe conocimiento a priori suficiente para reducir el espacio de soluciones.

---

### 4.2 Disponibilidad de datos

| Fuente de datos | Tipo | Disponibilidad | Sensibilidad |
|---|---|---|---|
| Registros históricos de triaje (HCE) | Tabular + texto | Alta (hospitales públicos) | Alta — RGPD |
| Constantes vitales en urgencias | Numérico en tiempo real | Alta (monitores conectados) | Media |
| Diagnósticos CIE-10 (codificados) | Categórico | Alta (sistemas hospitalarios) | Media |
| Descripciones de síntomas (texto libre) | NLP | Media (no siempre digitalizadas) | Alta — RGPD |
| Datasets públicos (MIMIC-IV, eICU) | Mixto | Alta (acceso regulado con firma) | Media-Alta |

Se estima que un hospital de tamaño medio genera **~100.000 episodios de urgencia/año**, suficiente para entrenamiento robusto con técnicas supervisadas.

**Consideración RGPD:** Los datos sanitarios son datos especialmente protegidos (Art. 9 RGPD). Se requerirá un **Data Protection Impact Assessment (DPIA)** y anonimización de los conjuntos de entrenamiento.

---

### 4.3 Análisis Coste-Beneficio

```python
# Análisis coste-beneficio estimado (€ aproximados para un hospital de tamaño medio)

categorias = ["Desarrollo\nIA + NLP", "Integración\nHCE", "Validación\nClínica",
               "Cumplimiento\nLey IA UE", "Formación\nPersonal", "Mantenimiento\nAnual"]
costes = [180_000, 60_000, 40_000, 25_000, 15_000, 30_000]  # € estimados

beneficios_cat = ["Reducción\ninternaciones\nevitables", "Ahorro tiempo\ntriaje\n(-60%)",
                  "Reducción\nerrores\npriorización", "Mejora\nsatisfacción\npaciente",
                  "Posicionamiento\nmercado\n(licitaciones)"]
beneficios_val = [250_000, 120_000, 95_000, 60_000, 200_000]  # € estimados anuales

total_coste = sum(costes)
beneficio_anual = sum(beneficios_val)
roi_anio1 = (beneficio_anual - total_coste) / total_coste * 100

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Costes
colores_coste = ['#e74c3c', '#e67e22', '#f39c12', '#d35400', '#c0392b', '#a93226']
wedges, texts, autotexts = axes[0].pie(
    costes, labels=categorias, colors=colores_coste,
    autopct='%1.1f%%', startangle=90, textprops={'fontsize': 7}
)
axes[0].set_title(f"Costes Totales\n({total_coste/1000:.0f}k €)", fontweight='bold')

# Beneficios
colores_benef = ['#27ae60', '#2ecc71', '#1abc9c', '#16a085', '#0e6655']
axes[1].pie(
    beneficios_val, labels=beneficios_cat, colors=colores_benef,
    autopct='%1.1f%%', startangle=90, textprops={'fontsize': 7}
)
axes[1].set_title(f"Beneficios Anuales Estimados\n({beneficio_anual/1000:.0f}k €/año)", fontweight='bold')

# ROI por años
anios = np.arange(1, 6)
beneficio_acumulado = beneficio_anual * anios
coste_acumulado = total_coste + (30_000 * anios)  # coste inicial + mantenimiento anual

axes[2].fill_between(anios, beneficio_acumulado/1000, coste_acumulado/1000,
                     where=(beneficio_acumulado > coste_acumulado),
                     alpha=0.3, color='green', label='Zona de beneficio neto')
axes[2].fill_between(anios, beneficio_acumulado/1000, coste_acumulado/1000,
                     where=(beneficio_acumulado <= coste_acumulado),
                     alpha=0.3, color='red', label='Zona de inversión')
axes[2].plot(anios, beneficio_acumulado/1000, 'g-o', linewidth=2, label='Beneficio acumulado')
axes[2].plot(anios, coste_acumulado/1000, 'r--s', linewidth=2, label='Coste acumulado')
axes[2].set_xlabel("Años desde implantación")
axes[2].set_ylabel("Miles de €")
axes[2].set_title("Proyección ROI (5 años)", fontweight='bold')
axes[2].legend(fontsize=7)
axes[2].grid(alpha=0.3)

# Punto de equilibrio
breakeven = total_coste / (beneficio_anual - 30_000)
axes[2].axvline(x=breakeven, color='purple', linestyle=':', linewidth=1.5)
axes[2].text(breakeven + 0.05, 100, f'Break-even\n~{breakeven:.1f} años',
             fontsize=8, color='purple')

plt.suptitle(f"Análisis Coste-Beneficio | ROI año 1: {roi_anio1:.0f}%",
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('analisis_coste_beneficio.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\nResumen financiero:")
print(f"  Inversión inicial total:   {total_coste:>10,.0f} €")
print(f"  Beneficio anual estimado:  {beneficio_anual:>10,.0f} €")
print(f"  ROI primer año:            {roi_anio1:>9.1f} %")
print(f"  Punto de equilibrio:       ~{breakeven:.1f} años")
```
